#### Train test split


```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
X=dataset.drop("prognosis",axis=1) # drop the last column for axis=1
y=dataset["prognosis"]
```

```python
X
```

```python
y
```

```python
le=LabelEncoder()
le.fit(y)
Y=le.transform(y)  # label encoding is just to label the string with some number
```

```python
Y
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,Y,test_size=0.3,random_state=20)
# this is method is just splitting the X and y into sets where X_train consits
# of 70% of the X while 30% goes  with the X_test same as happen with the y
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

#### Trainig the models


```python
from sklearn.datasets import make_classification
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB

from sklearn.metrics import accuracy_score,confusion_matrix
import numpy as np


# here I have create a dictionary to store models

models={
    "SVC":SVC(kernel='linear'),  # here linear because the dataset is linear
    "RandomForest":RandomForestClassifier(n_estimators=100,random_state=42), 
    "GradientBoosting":GradientBoostingClassifier(n_estimators=100,random_state=42),
    "KNeighbors":KNeighborsClassifier(n_neighbors=5),
    "Multinomial":MultinomialNB(),
}

for model_name,model in models.items():
    # train model
    model.fit(X_train,y_train)

    # test model
    predications=model.predict(X_test)

    # calculate accurracy
    accuracy=accuracy_score(y_test,predications)

    # calculate confusion matrix
    cm=confusion_matrix(y_test,predications)


    print(f"{model_name} accuracy: {accuracy}")
    print(f"{model_name} confusion matrix")
    print(np.array2string(cm,separator=', '))
```

#### single prediction


```python
svc=SVC(kernel='linear')
svc.fit(X_train,y_train)
ypred=svc.predict(X_test)
accuracy_score(y_test,ypred)
```

#### saving and loading the model


```python
# saving model
import pickle 
pickle.dump(svc,open("svc.pkl",'wb'))
```

```python
# loading the model
svc=pickle.load(open("svc.pkl",'rb'))
```

### just creating two dictionaries


```python
# create a simple dictionary of column names with their indices
def myfun(data):
    symptoms_dict={}
    index=0
    for i in data.columns.tolist():
        symptoms_dict[i]=index
        index+=1
    return symptoms_dict
    
```

```python
symptoms_dict=myfun(X)
```

```python
symptoms_dict
```

```python
# create a dictionary of label encoder with their disease name

def label_encoding_dict(data):
    label_encoder = LabelEncoder()
    encoded_labels = label_encoder.fit_transform(data)
    label_encoding_dict = {label: string for label, string in zip(encoded_labels, data)}
    return label_encoding_dict
```

### Recommendation model for the given symptoms


```python
## helper function
def helper(dis):
    desc=description_dataset[description_dataset["Disease"]==dis]["Description"]
    desc=" ".join([w for w in desc])


    pre=precautions_dataset[precautions_dataset["Disease"]==dis][["Precaution_1","Precaution_2","Precaution_3","Precaution_4"]]
    pre=[col for col in pre.values]
    

    med=medication_dataset[medication_dataset["Disease"]==dis]["Medication"]
    med=[med for med in med.values]


    diets=diets_dataset[diets_dataset["Disease"]==dis]['Diet']
    diets=[diets for diets in diets.values]

    workout=workout_dataset[workout_dataset["disease"]==dis]['workout']
    workout=[workout for workout in workout.values]

    return desc,pre,med,diets,workout
```

```python
## model prediction function
def get_predicted_value(patient_symptoms):
    input_vector=np.zeros(len(symptoms_dict))
    for item in patient_symptoms:
        input_vector[symptoms_dict[item]]=1
    return label_disease[svc.predict([input_vector])[0]]
```

```python
# test 1
symptoms=input("Enter your symptoms ....")
user_symptoms=[s.strip() for s in symptoms.split(',')]
user_symptoms=[sym.strip("[]' ")  for sym in user_symptoms]
predicted_disease=get_predicted_value(user_symptoms)
desc,pre,med,diets,workout=helper(predicted_disease)
```
